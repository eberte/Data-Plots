# The Tidyverse Package & Data Plots

We've been learning how to plot data and use the functionality and data built-in to the tidyverse package. These notes are meant to be a reference for you to remind yourself of the things that we have done in class. We begin by loading the tidyverse library below:

````r
library("tidyverse")
````

We began by creating a scatter plot of the size of the engine vs. the highway miles per gallon. We did this as follows:

````r
ggplot(data = mpg) + geom_point(mapping = aes(x = displ, y = hwy))
````

We were able to add some color, by class of the car, to give us some more information:

````r
ggplot(data = mpg) + geom_point(mapping = aes(x = displ, y = hwy, color = class))
````

We used the ```geom_smooth``` to plot a curve that "fits" the data. The parameter ```se = FALSE``` excludes information about confidence intervals from the graph.

````r
ggplot(data = mpg) + geom_smooth(mapping = aes(x = displ, y = hwy), se = FALSE)
````

Here we include the data points with our best fit curve. Plotting with ```ggplot``` is done in layers. This is why the '+' signs. Each layer is being added to the plot.

````r
ggplot(data = mpg, mapping = aes (x = displ, y = hwy)) + geom_point(mapping = aes(color = class)) + geom_smooth(se = FALSE)
````

We discussed how we believed the "2seater" class, all of which are Corvettes, were probably a larger influence on the data than desired. We proceeded to exclude this data, and do the plots without it. The methodology below is slightly different that what I did in class. However, I think this is probably more readable code, than simply hard removing certain lines from the data frame. It is certainly more general, and the technique can be used with any data frame.

````r
notTwoSeats <- mpg$class != "2seater"
newData <- mpg[notTwoSeats,]
ggplot(data = newData, mapping = aes(x = displ, y = hwy)) + geom_point(mapping = aes(color = class)) + geom_smooth(se = FALSE)
````

Here we have included all of the data points, but the curve is only fitted to the cars with the "subcompact" class. We did this through the use of filter.

````r
ggplot(data = mpg, mapping = aes(x = displ, y = hwy)) + geom_point(mapping = aes(color = class)) + geom_smooth(data = filter(mpg, class == "subcompact"), se = FALSE)
````
